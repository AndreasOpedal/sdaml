## Wiener filter the signals

```python
# Compute MSE of wiener filtered signal and original
mses = [np.array([]) for i in range(0,4)]
cs = [c0, c1, c2, c3]
for j in range(0, len(cs)):
    for i in range(1,10):
        orig = xtrain_clean.values[cs[j]][i]
        wt = scipy.signal.wiener(orig, mysize=None, noise=None)
        mse = sklearn.metrics.mean_squared_error(wt, orig)
        mses[j] = np.append(mses[j], mse)
```

## Now try to apply this to an svm one-vs-all classification

```python
cx2 = np.arange(len(ytrain.values))[np.arange(len(ytrain.values)) not in c3]
cx = cx2[0] # The indices of the non c3 classes
ytrain_c3 = ytrain.values.copy()
ytrain_c3 = np.ones_like(ytrain.values.copy())
ytrain_c3[c3] = 0
```

```python
# Make the X input the fourier transform
window = 50
X_ft = np.real(np.fft.rfft(xtrain_clean.values, axis=1))[:, 0:window] # Only take the real component of the output?

# Try to wiener filter features 
orig = xtrain_clean.values
wiener_mse = np.zeros(orig.shape[0])
for i in range(len(orig)):
    wt = scipy.signal.wiener(orig[i], noise=None)
    mse = sklearn.metrics.mean_squared_error(wt, orig[i])
    wiener_mse[i] = mse
wiener_mse = np.reshape(wiener_mse, (wiener_mse.shape[0], 1))
    
# Try to add standard deviation, mean, median as features
stds = np.std(xtrain_clean.values, axis=1)
stds = np.reshape(stds, (stds.shape[0], 1))
med = np.median(xtrain_clean.values, axis=1)
med = np.reshape(med, (med.shape[0], 1))
men = np.mean(xtrain_clean.values, axis=1)
men = np.reshape(men, (men.shape[0], 1))

# Stack all of the features
X_features = np.hstack((X_ft, wiener_mse, stds, med, men)) # The final X array of features
```

```python
kfold = KFold(n_splits=3, shuffle=True)
svm_model = svm.SVC(class_weight='balanced', degree=1, gamma='auto', kernel='rbf', probability=True)
results = cross_val_score(svm_model, X_features, ytrain_c3, cv=kfold, verbose=2)
print(results)
```

```python
# See whether it is classifying correctly
svm_model.fit(X_features[0:-1000], ytrain_c3[0:-1000])
predy = svm_model.predict(X_features[-1000:-1])
print(sklearn.metrics.classification_report(ytrain_c3[-1000: -1], predy))
```
